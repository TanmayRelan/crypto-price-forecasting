```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")

import keras
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.layers import Dense, LSTM, Dropout

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from datetime import timedelta

# Load the dataset
data_dir = 'BTC-INR.csv'
df = pd.read_csv(data_dir)

# Ensure Date is in datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Group by date and take the mean of the 'Close' price
df = df.groupby('Date')['Close'].mean()
df = pd.DataFrame(df)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(np.array(df.values).reshape(-1, 1))

# 75% to Train, 25% to Test
train_size = int(len(df) * 0.75)
test_size = len(df) - train_size

print("Train Size:", train_size, "Test Size:", test_size)

train_data = scaled_data[:train_size, 0:1]
test_data = scaled_data[train_size-60:, 0:1]  # Note: test_data starts from train_size-60 to ensure continuity

# Creating a Training set with 60 time-steps and 1 output
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

# Convert to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape for LSTM [samples, time-steps, features]
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

# Build the LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(x_train.shape[1], 1)),
    Dropout(0.2),  # Add dropout to prevent overfitting
    LSTM(64, return_sequences=False),
    Dropout(0.2),  # Add dropout to prevent overfitting
    Dense(32),
    Dense(16),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mean_absolute_error'])

# Summary of the model
model.summary()

# Early stopping to avoid overfitting
callbacks = [EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)]

# Train-test split for validation
x_train_split, x_val_split, y_train_split, y_val_split = train_test_split(x_train, y_train, test_size=0.2, shuffle=False)

# Fit the model
history = model.fit(x_train_split, y_train_split, validation_data=(x_val_split, y_val_split), epochs=100, batch_size=32, callbacks=callbacks)

# Plotting training & validation loss and MAE
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss over Epochs')

plt.subplot(1, 2, 2)
plt.plot(history.history['mean_absolute_error'], label='Train MAE')
plt.plot(history.history['val_mean_absolute_error'], label='Validation MAE')
plt.legend()
plt.title('Mean Absolute Error over Epochs')

plt.show()

# Testing on unseen data
x_test = []
y_test = []

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
    y_test.append(test_data[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

# Predicting on test data
predictions = model.predict(x_test)

# Inverse scaling
predictions = scaler.inverse_transform(predictions)
y_test = scaler.inverse_transform([y_test])

# Calculate RMSE
RMSE = np.sqrt(mean_squared_error(y_test[0], predictions)).round(2)
print(f"Test RMSE: {RMSE}")

# Visualizing the predictions vs actual prices
train = df.iloc[:train_size]
test = df.iloc[train_size:]
test['Predictions'] = predictions

plt.figure(figsize=(16, 6))
plt.title('Bitcoin Price Prediction', fontsize=20)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Price in INR', fontsize=18)
plt.plot(train['Close'], label='Train')
plt.plot(test['Close'], label='Test')
plt.plot(test['Predictions'], label='Predicted')
plt.legend()
plt.show()

def insert_end(Xin, new_input):
    timestep = Xin.shape[1]  # The number of time steps in the input
    for i in range(timestep - 1):
        Xin[:, i, :] = Xin[:, i + 1, :]
    Xin[:, timestep - 1, :] = new_input
    return Xin

# Forecasting future prices for the next 30 days
future = 30
forecast = []
Xin = x_test[-1:, :, :]
time = []

for i in range(future):
    out = model.predict(Xin, batch_size=1)
    forecast.append(out[0, 0])
    Xin = insert_end(Xin, out[0, 0])
    time.append(df.index[-1] + timedelta(days=i))

# Convert forecast to DataFrame
forecasted_output = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))
df_forecast = pd.DataFrame({'Date': time, 'Forecasted': forecasted_output.flatten()})

plt.figure(figsize=(16, 6))
plt.title('Bitcoin Forecast for Next 30 Days', fontsize=20)
plt.xlabel('Date', fontsize=18)
plt.ylabel('Price in INR', fontsize=18)
plt.plot(df_forecast['Date'], df_forecast['Forecasted'], marker='o', color='red', label='Forecasted Price')
plt.legend()
plt.show()
```
